# TRANSFORM

### 1st Dataset: Median Income by State (1984 - 2019)
 - Drop columns (standard error and all years prior to 2015)
 - Rename columns
 - Remove irrelevant rows by index
 - Reset index
 - Remove commas in median income values
 - Convert median income values from string to integer
 - Check for missing values

```python
median_income_df.head()
```

***
<font color = blue>__Check column names__</font>

```python
median_income_df.columns
```

***
<font color = blue>__Remove all standard error columns__</font>

```python
median_income_df = median_income_df.drop(columns=["Unnamed: 2", "Unnamed: 4", "Unnamed: 6", "Unnamed: 8",\
                                                 "Unnamed: 10", "Unnamed: 12", "Unnamed: 14", "Unnamed: 16",\
                                                 "Unnamed: 18", "Unnamed: 20", "Unnamed: 22", "Unnamed: 24",\
                                                 "Unnamed: 26", "Unnamed: 28", "Unnamed: 30", "Unnamed: 32",\
                                                 "Unnamed: 34", "Unnamed: 36", "Unnamed: 38", "Unnamed: 40",\
                                                 "Unnamed: 42", "Unnamed: 44", "Unnamed: 46", "Unnamed: 48",\
                                                 "Unnamed: 50", "Unnamed: 52", "Unnamed: 54", "Unnamed: 56",\
                                                 "Unnamed: 58", "Unnamed: 60", "Unnamed: 62", "Unnamed: 64",\
                                                 "Unnamed: 66", "Unnamed: 68", "Unnamed: 70", "Unnamed: 72",\
                                                 "Unnamed: 74", "Unnamed: 76"])
median_income_df.head()
```

***
<font color = blue>__View remaining column names__</font>

```python
median_income_df.columns
```

***
<font color = blue>__Drop all years prior to 2015__</font>

```python
median_income_df = median_income_df.drop(columns=["1984 (19)", "1985 (20)", "1986", "1987 (21)", "1988",\
                                                 "1989", "1990", "1991", "1992 (22)", "1993 (23)", "1994 (24)",\
                                                 "1995 (25)", "1996", "1997", "1998", "1999 (29)", "2000 (30)",\
                                                 "2001", "2002", "2003", "2004 (revised)", "2005", "2006", "2007",\
                                                 "2008", "2009 (36)", "2010 (37)", "2011", "2012", "2013 (38)",\
                                                 "2013 (39)", "2014", "2017"])
median_income_df.head()
```

***
<font color = blue>__Rename columns. Med. Income = Median Income__</font>

```python
median_income_df = median_income_df.rename(columns = {"2019": "2019 Med. Income", "2018": "2018 Med. Income",\
                                                      "2017 (40)": "2017 Med. Income", "2016": "2016 Med. Income",\
                                                      "2015": "2015 Med. Income"})
median_income_df.head()
```

***
<font color = blue>__Remove first 2 rows__</font>

```python
median_income_df = median_income_df.drop([0, 1])
median_income_df.head()
```

***
<font color = blue>__Reset index & drop index__</font>

```python
median_income_df = median_income_df.reset_index(drop=True)
median_income_df
```

***
<font color = blue>__View dataframe's datatypes__</font>

```python
median_income_df.dtypes
```

***
<font color = blue>__Remove commas from median income values__</font>

```python
median_income_df = median_income_df.replace(',','', regex= True)
median_income_df.head()
```

***
<font color = blue>__Convert median income values (strings) to integers__</font>

```python
median_income_df[['2019 Med. Income', '2018 Med. Income', '2017 Med. Income', '2016 Med. Income', '2015 Med. Income']]\
= median_income_df[['2019 Med. Income', '2018 Med. Income', '2017 Med. Income', '2016 Med. Income', '2015 Med. Income']].apply(pd.to_numeric)

print(median_income_df.dtypes)
```

***
<font color = blue>__Confirm no missing values__</font>

```python
median_income_df.isna().sum()
```

### 2nd Dataset: Population by State (2010 - 2019)
 - Drop columns (all years prior to 2015)
 - Rename columns
 - Reorder columns
 - Remove irrelevant rows by index
 - Remove . in state column
 - Reset index
 - Remove commas in population figures
 - Convert population figures from string to integer
 - Check for missing values

```python
population_df.head()
```

***
<font color = blue>__Check column names__</font>

```python
population_df.columns
```

***
<font color = blue>__Remove columns prior to 2015__</font>

```python
population_df = population_df.drop(columns=["April 1, 2010", "Unnamed: 2", "Population Estimate (as of July 1)",\
                                            "Unnamed: 4", "Unnamed: 5", "Unnamed: 6", "Unnamed: 7"])
population_df.head()
```

***
<font color = blue>__Rename columns__</font>

```python
population_df = population_df.rename(columns = {"Geographic Area": "State", "Unnamed: 8": "2015 Population",\
                                                "Unnamed: 9": "2016 Population", "Unnamed: 10": "2017 Population",\
                                                "Unnamed: 11": "2018 Population", "Unnamed: 12": "2019 Population"})
population_df.head()
```

***
<font color = blue>__Reorder columns__</font>

```python
population_df = population_df[['State', '2019 Population', '2018 Population', '2017 Population', '2016 Population', '2015 Population']]
population_df
```

***
<font color = blue>__Drop irrelevant rows (first 6 rows)__</font>

```python
population_df = population_df.drop([0, 1, 2, 3, 4, 5])
population_df.head(10)
```

***
<font color = blue>__Remove . in front of state names__</font>

```python
population_df['State'] = population_df['State'].str.replace('.','')
population_df.head()
```

***
<font color = blue>__Reset index & drop index__</font>

```python
population_df = population_df.reset_index(drop=True)
population_df.head()
```

***
<font color = blue>__View dataframe's datatypes__</font>

```python
population_df.dtypes
```

***
<font color = blue>__Remove commas from population figures__</font>

```python
population_df = population_df.replace(',','', regex= True)
population_df.head()
```

***
<font color = blue>__Convert population figures from strings to integers__</font>

```python
population_df[['2019 Population', '2018 Population', '2017 Population', '2016 Population', '2015 Population']] = \
population_df[['2019 Population', '2018 Population', '2017 Population', '2016 Population', '2015 Population']].apply(pd.to_numeric)

print(population_df.dtypes)
```

***
<font color = blue>__Confirm no missing values__</font>

```python
population_df.isna().sum()
```
