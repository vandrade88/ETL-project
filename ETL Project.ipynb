```python
median_income_df = median_income_df.reset_index(drop=True)
median_income_df
```

***

```python
median_income_df.dtypes
```

<font color = blue>__Remove commas from median income values__</font>

```python
median_income_df = median_income_df.replace(',','', regex= True)
median_income_df.head()
```

***

### Merge 5 Unemployment Rate Dataframes into 1

```python
unemp_1 = pd.merge(unemp_2019_df, unemp_2018_df, on="State")
unemp_1.head()
```

```python
unemp_2 = pd.merge(unemp_1, unemp_2017_df, on="State")
unemp_2.head()
```

```python
unemp_3 = pd.merge(unemp_2, unemp_2016_df, on="State")
unemp_3.head()
```

```python
unemp_final_df = pd.merge(unemp_3, unemp_2015_df, on="State")
unemp_final_df.head()
```
